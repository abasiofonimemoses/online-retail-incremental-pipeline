```python
import pandas as pd

file_path = "/Volumes/workspace/default/retail_volume/Online Retail.xlsx"

pdf = pd.read_excel(file_path)

print(pdf.shape)
display(pdf.head())
```

```python
print(pdf.head().to_string())
```

```python
print(pdf.columns)
print(pdf.dtypes)
```

```python
import os

base_path = "/Volumes/workspace/default/retail_volume"
archives_path = f"{base_path}/archives"

os.makedirs(archives_path, exist_ok=True)

pdf["InvoiceDate"] = pd.to_datetime(pdf["InvoiceDate"])
pdf["year"] = pdf["InvoiceDate"].dt.year
pdf["month"] = pdf["InvoiceDate"].dt.month

for (year, month), month_df in pdf.groupby(["year", "month"]):
    file_name = f"sales_{year}_{month:02d}.csv"
    output_path = f"{archives_path}/{file_name}"
    month_df.drop(columns=["year", "month"]).to_csv(output_path, index=False)
    print(f"Created {file_name}: {len(month_df)} rows")

print("Dataset partitioning completed.")
```
