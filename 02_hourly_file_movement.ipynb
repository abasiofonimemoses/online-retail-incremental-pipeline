```python
import os
import shutil

base_path = "/Volumes/workspace/default/retail_volume"
archives_path = f"{base_path}/archives"
landing_path = f"{base_path}/landing"

os.makedirs(landing_path, exist_ok=True)

print("Archives path:", archives_path)
print("Landing path:", landing_path)
```

```python
archive_files = sorted([f.name for f in dbutils.fs.ls(archives_path) if f.name.endswith(".csv")])
landing_files = sorted([f.name for f in dbutils.fs.ls(landing_path) if f.name.endswith(".csv")])

pending_files = [file for file in archive_files if file not in landing_files]

print("Total archive files:", len(archive_files))
print("Files already in landing:", len(landing_files))
print("Pending files:", len(pending_files))

if pending_files:
    next_file = pending_files[0]
    source = f"{archives_path}/{next_file}"
    destination = f"{landing_path}/{next_file}"

    dbutils.fs.cp(source, destination)
    print(f"Moved one file to landing: {next_file}")
else:
    print("No pending files left to move.")
```
